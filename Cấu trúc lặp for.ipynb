## Review: Vẽ kim tự tháp


```python
# Nhập vào chiều cao kim tự tháp:
import turtle
height = int(input("Nhập chiều cao của kim tự tháp: "))
t = turtle.Turtle()
count = 1
while count <= height:
    t.circle(count*20, steps=3)
    count+=1
    
```

```python
turtle.done()
```

## for kết hợp với hàm range()

### Trường hợp 1 tham số

```python
count = 0
while count <= 10:
    print(count)
    count+=1
print("---")
for i in range(11):
    print(i)
```

### Ví dụ: In các số lẻ từ 0 đến 100

```python
for i in range(101):
    if i%2==1:
        print(i)
```

### Trường hợp 2 tham số

```python
for i in range(50,101):
    if not i%2:
        print(i)
```

### Ví dụ: In các số chia hết cho 7 từ a đến b

```python
a, b = int(input("Nhập vào số a: ")), int(input("Nhập vào số b: "))
if a<b:
    for i in range(a,b+1):
        if not i%7:
            print(i)
```

### Trường hợp 3 tham số

```python
for i in range(0,101,5):
    print(i)
```

```python
for i in range(101,0,-5):
    print(i)
```

### Ví dụ: Biểu diễn cấp số cộng

```python
# Viết chương trình nhập vào số hạng đầu(n0) và công sai(d) của 1 cấp số cộng.
# In ra 10 số hạng tiếp theo của cấp số cộng đó!
n0 = int(input("Nhập số hạng đầu: "))
d = int(input("Nhập công sai: "))
for i in range(n0+d,n0+11*d,d):
    print(i)
```

## lặp qua một chuỗi

```python
content = "This is a nice day!"
for c in content:
    print(c)
```

### Ví dụ: Đếm số lượng nguyên âm trong chuỗi

```python
chuoi = input("Hãy nhập 1 chuỗi: ")
count = 0
for c in chuoi:
    if c in "aeiou":
        count+=1
print(count)
```

## Lặp qua một danh sách chuỗi

```python
names = "Long Huy Nam Khánh Minh Kiên"
name_list = names.split()
print(name_list)
for member in name_list:
    print(member)
```

### Ví dụ: Tìm chuỗi dài nhất

```python
max_len = 0
for member in name_list:
    if len(member) > max_len:
        max_len = len(member)
print(f"Tên dài nhất có {max_len} ký tự")
```

## Vòng lặp lồng nhau

```python
clothes = ["shirt", "pants", "sweater"]
colors = ["blue", "black", "yellow"]
for item in clothes:
    for color in colors:
        print(f"{colors} {item}")
```
