```python
# Import Turtle
import turtle
t = turtle.Turtle()
turtle.speed(10)
```

```python
#Vẽ cỏ p1
mau_co = "#0AA344"
t.penup()
t.goto(-320,-10)
t.fillcolor(mau_co)
t.begin_fill()
t.pendown()
t.fd(300)
t.rt(90)
t.fd(250)
t.rt(90)
t.fd(300)
t.end_fill()
```

```python
# Vẽ trăng
mau_trang1 = "#45B6FE"
mau_trang2 = "#E6E9ED"
t.pencolor(mau_troi)
t.penup()
t.goto(280,200)
t.pendown()
t.fillcolor(mau_trang1)
t.begin_fill()
t.circle(60)
t.end_fill()

t.pencolor(mau_trang1)
t.penup()
t.goto(265,200)
t.pendown()
t.fillcolor(mau_trang2)
t.begin_fill()
t.circle(40)
t.end_fill()
t.pencolor("black")
```
