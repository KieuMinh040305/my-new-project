```python
import turtle
pen = turtle.Turtle()
```

```python
pen.pensize(5)
```

```python
# Ve Mat Troi ne!
pen.penup()
pen.goto(140,90)
pen.pencolor("orange")
pen.fillcolor("yellow")
pen.begin_fill()
pen.pendown()
pen.circle(50)
pen.end_fill()

pen.right(90)
pen.pencolor("yellow")
pen.forward(25)

pen.penup()
pen.goto(90, 140)
pen.right(90)
pen.pendown()
pen.forward(25)

pen.penup()
pen.goto(190,140)
pen.right(180)
pen.pendown()
pen.forward(25)

pen.penup()
pen.goto(140, 190)
pen.left(90)
pen.pendown()
pen.forward(25)

pen.penup()
pen.goto(105, 175)
pen.left(45)
pen.pendown()
pen.forward(15)

pen.penup()
pen.goto(175,175)
pen.right(90)
pen.pendown()
pen.forward(15)

pen.penup()
pen.goto(175,105)
pen.right(90)
pen.pendown()
pen.forward(15)

pen.penup()
pen.goto(105, 105)
pen.right(90)
pen.pendown()
pen.forward(15)
```
