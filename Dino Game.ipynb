```python
from tkinter import *
from PIL import ImageTk, Image
from playsound import playsound

# Thiết lập cơ bản
game = Toplevel()
game.title("Dino game")
game.geometry("700x350")
canvas = Canvas(master = game, height = 300, width = 600, bg = "white")
canvas.pack()

# Hình ảnh
img1 = PhotoImage(file = "dragon.png")
img2 = PhotoImage(file = "cloud.png")
img3 = PhotoImage(file = "tree2.png")
img4 = PhotoImage(file = "land5.png")

# Gắn vào Canvas trên Tkinter
land_image = canvas.create_image(0 , 250, anchor=NW, image=img4)
dinosaur_image = canvas.create_image(0, 250, anchor=NW, image=img1)
cloud_image = canvas.create_image(550, 60, anchor=NW, image=img2)
cloud_image2 = canvas.create_image(450, 90, anchor=NW, image=img2)
cloud_image3 = canvas.create_image(300, 60, anchor=NW, image=img2)
tree_image = canvas.create_image(550, 250, anchor=NW, image=img3)

# Mây
def cloud_move():
    global cloud_image
    canvas.move(cloud_image, -5, 0)
    if canvas.coords(cloud_image) == [0.0, 60.0]:
        canvas.delete(cloud_image)
        cloud_image = canvas.create_image(550, 60, anchor=NW, image=img2)

def cloud_move2():
    global cloud_image2
    canvas.move(cloud_image2, -5, 0)
    if canvas.coords(cloud_image2) == [0.0, 90.0]:
        canvas.delete(cloud_image2)
        cloud_image2 = canvas.create_image(450, 90, anchor=NW, image=img2)

def cloud_move3():
    global cloud_image3
    canvas.move(cloud_image3, -5, 0)
    if canvas.coords(cloud_image3) == [0.0, 60.0]:
        canvas.delete(cloud_image3)
        cloud_image3 = canvas.create_image(300, 60, anchor=NW, image=img2)


# Cây
def tree_move():
    global tree_image, score
    canvas.move(tree_image, -5, 0)
    if canvas.coords(tree_image) == [0.0, 250.0]:
        canvas.delete(tree_image)
        tree_image = canvas.create_image(550, 250, anchor=NW, image=img3)
        score += 1
        canvas.itemconfig(score_text, text = "Score " + str(score))

# Điểm và hướng dẫn
score = 0 
score_text = canvas.create_text(535, 55, text = "Score " + str(score), font = ("Times", 8), fill = "Black")
rule_text = canvas.create_text(535, 25, text = "Space để bắt đầu", font = ("Times", 8), fill = "Black")
rule_text2 = canvas.create_text(535, 10, text = "Up để nhảy", font = ("Times", 8), fill = "Black")
rule_text2 = canvas.create_text(535, 40, text = "Space nhiều lần để tăng tốc", font = ("Times", 8), fill = "Black")

# Nhảy
def jump_down():    
    for i in range(0, 20):
        canvas.move(dinosaur_image, 0, 5)
def jump(e):
    for i in range(0, 20):
        canvas.move(dinosaur_image, 0, -5)
    game.after(800, jump_down)
game.bind("<Up>", jump)
gameover = False     

def move_object():
    cloud_move()
    cloud_move2()
    cloud_move3()
    tree_move() 
def process_game():
    global gameover
    if not gameover:
        move_object()
        gameover_check()
        game.after(100, process_game)
def game_start(e):
    process_game()
game.bind("<space>", game_start)

# Va chạm dẫn tới gameover
def gameover_check():
    global gameover
    if canvas.coords(tree_image) < [50.0, 250.0] and canvas.coords(dinosaur_image) > [0.0, 220.0]:
        gameover_text = canvas.create_text(250, 150, text = "Game Over", font = ("Times", 30), fill = "Black")
        gameover = True
        
game.mainloop()
```
